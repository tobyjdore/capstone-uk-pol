```python
auth = tweepy.OAuthHandler(creds['CONSUMER_KEY'], creds['CONSUMER_SECRET'])
auth.set_access_token(creds['ACCESS_TOKEN'], creds['ACCESS_SECRET'])

api = tweepy.API(auth)

tweets = {'date':[],
          'party':[],
          'text':[]}
for party in ['UKLabour','Conservatives']:
    for status in tweepy.Cursor(api.user_timeline, screen_name=party, tweet_mode="extended").items():
        try:
            tweets['text'].append(status.retweeted_status.full_text)
            tweets['party'].append(party)
            tweets['date'].append(status.retweeted_status.created_at)
        except:
            tweets['text'].append(status.full_text)
            tweets['party'].append(party)
            tweets['date'].append(status.created_at)
```

```python
import pandas as pd
import numpy as np
import ast

from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
```

```python
tweets_df = pd.DataFrame(tweets)
```

```python
tweets_df.head()
```

```python
tweets_df.party = tweets_df.party.map(lambda x: 'Labour' if x=='UKLabour' else 'Conservative')
```

```python
tweets_df.party.value_counts()
```

```python
uk_pol_tokens = pd.read_csv('uk_pol_tokens.csv')
```

```python
stop = stopwords.words('english')
# add additional stopwords based on the strongest predictors from the weak model:
stop = stop + ['conservative',
               'conservatives',
               'tory','tories',
               'labour','jeremy',
               'corbyn','george',
               'may','pdf','.']
```

```python
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings(action='ignore', category=ConvergenceWarning)
```

### Time to see if the model trained on the speeches can predict the party based on the tweets:

```python
pipeline = Pipeline([
    ('vect', CountVectorizer(lowercase=True, strip_accents='unicode', stop_words=stop)),
    ('logreg', LogisticRegression(penalty='l2', C=0.1, solver='lbfgs'))
]) 

pipeline.fit(uk_pol_tokens.tokenised, uk_pol_tokens.party)

print(pipeline.score(uk_pol_tokens.tokenised, uk_pol_tokens.party))
print(cross_val_score(pipeline, uk_pol_tokens.tokenised, uk_pol_tokens.party, cv=5).mean())
print(pipeline.score(tweets_df.text, tweets_df.party))

predictions = pipeline.predict(tweets_df.text)

print()
print(classification_report(tweets_df.party, predictions))

pd.DataFrame(confusion_matrix(tweets_df.party, predictions,
                              labels=['Conservative','Labour']),
             columns=['Conservative','Labour'],
             index=['Conservative','Labour'])
```

```python
vect = pipeline.named_steps['vect']
logreg = pipeline.named_steps['logreg']

coef_df = pd.DataFrame({'word':vect.get_feature_names(),'coef':logreg.coef_.flatten()})

coef_df['abs_coef'] = np.abs(coef_df.coef)

tory_coefs = coef_df.sort_values('coef',ascending=True).head(20)
labour_coefs = coef_df.sort_values('coef',ascending=False).head(20)

tory_coefs['factor'] = tory_coefs.coef.map(lambda x: -int(100*x))
labour_coefs['factor'] = labour_coefs.coef.map(lambda x: int(100*x))

tory_coef_list = []
for ind, row in tory_coefs.iterrows():
    for i in range(row.factor):
        tory_coef_list.append(row.word)
        
labour_coef_list = []
for ind, row in labour_coefs.iterrows():
    for i in range(row.factor):
        labour_coef_list.append(row.word)

def blue_color_func(word, font_path, font_size, position, orientation, random_state=None):
    return "#0087DC"

def red_color_func(word, font_path, font_size, position, orientation, random_state=None):
    return "#DC241f"

tory_coef_words = ' '.join(tory_coef_list)
labour_coef_words = ' '.join(labour_coef_list)
```

```python
tory_coef_wordcloud = WordCloud(collocations=False, color_func=blue_color_func, width=800, background_color='white').generate(tory_coef_words)
plt.figure(figsize=(20,10))
plt.imshow(tory_coef_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
labour_coef_wordcloud = WordCloud(collocations=False, color_func=red_color_func, width=800, background_color='white').generate(labour_coef_words)
plt.figure(figsize=(20,10))
plt.imshow(labour_coef_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
