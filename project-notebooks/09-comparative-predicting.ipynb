```python
import pandas as pd
import numpy as np
import ast
import matplotlib.pyplot as plt
import seaborn as sns
 
plt.style.use('ggplot')
%config InlineBackend.figure_format = 'retina'
%matplotlib inline

from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
```

```python
uk_pol_tokens = pd.read_csv('uk_pol_tokens.csv')
```

```python
uk_pol_tokens.head()
```

```python
type(uk_pol_tokens.iloc[0,4])
```

```python
stop = stopwords.words('english')
# add additional stopwords based on the strongest predictors from the weak model:
stop = stop + ['conservative','conservatives','tory','tories','labour','jeremy','corbyn','george','may','pdf','.']
```

```python
speakers = uk_pol_tokens.speaker.unique()
```

```python
output = pd.DataFrame()
```

```python
pipeline = Pipeline([
    ('vect', CountVectorizer(lowercase=True, strip_accents='unicode', stop_words=stop)),
    ('logreg', LogisticRegression(penalty='l2', C=0.1, solver='lbfgs'))
]) 

scores = {}

for name in speakers:
    
    train_df = uk_pol_tokens[uk_pol_tokens.speaker!=name]
    test_df = uk_pol_tokens[uk_pol_tokens.speaker==name]
    
    pipeline.fit(train_df.tokenised, train_df.party)
    score = cross_val_score(pipeline, train_df.tokenised, train_df.party, cv=3).mean()
    predictions = pipeline.predict(test_df.tokenised)
    
    scores[name]=score
    test_df['predictions']=predictions
    
    output = pd.concat((output, test_df))
```

```python
output.dropna(inplace=True)
```

```python
output[output.speaker=='Tony Blair']
```

```python
output[output.speaker=='Jeremy Corbyn']
```

```python
output['match'] = output.apply(lambda x: 1 if x.party==x.predictions else 0, axis=1)
```

```python
output.to_csv('comp_name.csv', index=False)
```

```python
scores_df = pd.DataFrame.from_dict(scores, orient='index', columns=['score'])
```
